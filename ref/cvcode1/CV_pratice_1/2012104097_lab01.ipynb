```
# derivative.py
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
filename = 'lena.bmp'
img = plt.imread(filename)


# Make derivative filter
filter_x = np.array([[-1, 0, 1]])
filter_y = np.array([-1,
                     0,
                     1])


# Apply filter to image
dst_x = cv2.filter2D(img, -1, filter_x)
dst_y = cv2.filter2D(img, -1, filter_y)

# Plot images
plt.subplot(231), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(232), plt.imshow(filter_x), plt.title('filter_x')
plt.xticks([]), plt.yticks([])
plt.subplot(233), plt.imshow(filter_x.T), plt.title('filter_y')
plt.xticks([]), plt.yticks([])

plt.subplot(235), plt.imshow(dst_x, cmap='gray'), plt.title('derivative_x')
plt.xticks([]), plt.yticks([])
plt.subplot(236), plt.imshow(dst_y, cmap='gray'), plt.title('derivative_y')
plt.xticks([]), plt.yticks([])

plt.show()
```

```
# gaussian_filter.py
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read image
filename = 'lena.bmp'
img = plt.imread(filename)


# Get gaussian filter
filter = cv2.getGaussianKernel(ksize=5, sigma=1)


# Make 2D gaussian filter ( Option )
filter = filter * filter.T


# Apply filter to image
dst = cv2.filter2D(img, -1, filter)


# Plot images
plt.subplot(131), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(dst, cmap='gray'), plt.title('Gaussian_blur')
plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(filter), plt.title('gaussian')
plt.xticks([]), plt.yticks([])
plt.show()
```

```
# laplacian_filter.py
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
filename = 'lena.bmp'
img = plt.imread(filename)


# Make laplacian filter
filter = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])


# Apply filter to image
dst = cv2.filter2D(img, -1, filter)


# Plot images
plt.subplot(131), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(dst, cmap='gray'), plt.title('Laplacian_filter')
plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(filter), plt.title('laplacian')
plt.xticks([]), plt.yticks([])
plt.show()
```
